**Analyzing bolwer economy**

```python
overs=ball.groupby('Bowler_Id')['Over_Id'].count()/6
```

```python
bowl_Overs=overs.reset_index()
```

```python
bowl_Overs.columns=['Bowler_Id','Ovrs_bowled']
bowl_Overs
```

```python
r_c=ball.groupby('Bowler_Id')['Batsman_Scored'].sum().reset_index()
```

```python
r_c
```

```python
bowl=pd.merge(bowl_Overs,r_c)
```

```python
bowl.head(2)
```

```python
bowl['eco']=(bowl['Batsman_Scored']/bowl['Ovrs_bowled']).round(2)
```

```python
atleast10ovrs=bowl[bowl['Ovrs_bowled']>10]
```

```python
atleast10ovrs.sort_values('eco')
```
